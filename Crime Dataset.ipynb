## WE HAVE REMOVED UNNECESSARY COLUMNBS

```python
dt = data.drop(['Date Reported', 'Date Occurred','Area Name','Crime Code Description','MO Codes','Victim Descent','Premise Description',
              'Weapon Description','Status Description', 'Crime Code 2', 'Crime Code 3','Crime Code 4', 'Address', 'Cross Street','Location ','Status Code', 'Victim Sex'],axis="columns")
```

```python
dt.head(20)
```

## REMOVEING NAN VALUES

```python
copy1 = dt.fillna(0)
copy1.head()
```

```python
copy1 = np.array(copy1)
```

```python
train_size = math.ceil(int(len(copy1))*0.5)
print(train_size)
test_size = math.ceil(int(len(copy1))*0.3)
print(test_size)
validation_data_size = math.ceil(int(len(copy1))*0.2)
print(validation_data_size)
792158+475295+316864
```

```python
train_data = copy1[:792158].astype("float64")
test_data = copy1[792158:1267453].astype('float64')
validation_data = copy1[1267453:1584317].astype('float64')
```

```python
train_label = copy1[:792158, 4:5]
print(train_label)
test_label = copy1[:test_size,8:9]
print(test_label)
validation_label = copy1[:validation_data_size,8:9]
print(validation_label)
```

```python

model = models.Sequential()
model.add(layers.Dense(10, activation='relu', input_shape=(9,)))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(1))
#model.compile(optimizer='rmsprop',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
```

```python
history = model.fit(train_data,
train_label,
epochs=10,
batch_size=512
        )
```

```python
results = model.evaluate(test_data, test_label)
```

```python

history = model.fit(train_data,
train_label,
epochs=100,
batch_size=4,
validation_data= (validation_data,validation_label)        )
```

```python
history_dict_v = history.history
history_dict_v.keys()
```

```python
import matplotlib.pyplot as plt
loss = history_dict_v['loss']
accuracy = history_dict_v['mae']
plt.plot(loss,accuracy, label='Training loss')
plt.title('Actual Training  loss')
plt.xlabel('Loss')
plt.ylabel('accuracy')
plt.legend()

plt.show() 
```
